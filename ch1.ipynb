### 페이지의 URL 정보 추출하기

```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
# 크롤링할 사이트 주소 정의
soure_url = "https://namu.wiki/RecentChanges"
```

```python
# 사이트의 HTML 구조에 기반하여 크롤링 수행
req = requests.get(soure_url) # url 의 html 문서 가져오기
html = req.content
soup = BeautifulSoup(html, 'lxml')    # Soup객체로 
contents_table = soup.find(name = "table")
table_body = contents_table.find(name ="tbody")
table_rows = table_body.find_all(name="tr")
```

```python
# a 태그의 href 속성을 리스트로 추출하여 크롤링할 페이지 리스트를 생성
page_url_base = "https://namu.wiki"
page_urls = []
for index in range(0, len(table_rows)):
    first_td = table_rows[index].find_all('td')[0]
    td_url = first_td.find_all('a')
    if len(td_url) > 0 :
        page_url = page_url_base + td_url[0].get('href')
        page_urls.append(page_url)
```

```python
# 중복 url 제거
page_urls = list(set(page_urls))
for page in page_urls[:5]:
    print(page)
```

### URL 페이지 정보를 기반으로 크롤링 하기

```python
# 하나의 최근 변경된 문서 크롤링
req = requests.get(page_urls[0])
html = req.content
soup = BeautifulSoup(html, 'lxml')
contents_table = soup.find(name ="article")
title = contents_table.find_all('h1')[0]       # 제목
category = contents_table.find_all('ul')[0]    # 카테고리
content_paragraphs = contents_table.find_all(name = "div")
content_corpus_list = []
```

```python
# 크롤링한 문서 정보 출력
for paragraphs in content_paragraphs:
    content_corpus_list.append(paragraphs.text) # 태그의 텍스트 정보만 추출
content_corpus = "".join(content_corpus_list)   # 합치기
```

### BeautifulSoup 이용해 웹 크롤링 하기

```python
# -*- coding: utf-8 -*-
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup
import re

# 크롤링할 사이트 주소 정의
soure_url = "https://namu.wiki/RecentChanges"

# 사이트 html 구조에 기반하여 크롤링 수행
req = requests.get(soure_url)
html = req.content
soup = BeautifulSoup(html, 'lxml')
contents_table = soup.find(name = "table")
크롤링할 사이트 주소 정의
soure_url = "https://namu.wiki/RecentChanges"

# 사이트 html 구조에 기반하여 크롤링 수행
req = requests.get(soure_url)
html = req.content
soup = BeautifulSoup(html, 'lxml')
contents_table = soup.find(name = "table")
table_body = contents_table.find(name = "tbody")
table_rows = table_body.find_all(name = "tr")
```

```python
# a 태그와 href 속성을 리스트로 추출하여, 크롤링할 페이지 리스트 생성
page_url_base = "https://namu.wiki"
page_urls = []
for index in range(0, len(table_rows)):
    first_td = table_rows[index].find_all('td')[0]
    td_url = first_td.find_all('a')
    if len(td_url) > 0 :
       page_url =page_url_base +td_url[0].get('href')
       if 'png' not in page_url:
           page_urls.append(page_url)

# 중복 url 제거하기
page_urls = list(set(page_urls))
```

### 나무위키의 최근 변경 데이터 크롤링하기

```python
# 크롤링한 데이터를 데이터 프레임으로 만들기 위한 준비
columns = ['title', 'content_text']
df = pd.DataFrame(columns=columns)
```

```python
# 각 페이지별 '제목', '카테고리', '본문' 정보를 데이터 프레임으로 만듬
for page_url in page_urls:

    # 사이트 html 구조에 기반하여 크롤링 수행
    req = requests.get(page_url)
    html = req.content
    soup = BeautifulSoup(html, 'lxml')

    contents_table = soup.find(name ="article")
    title = contents_table.find_all('h1')[0]       # 제목
    content_paragraphs = contents_table.find_all(name = "div",
    attr = {"class" : "wiki-paragraph"})
    content_corpus_list = []

    # 페이지 내 제목 정보에서 개행 문자를 제거한 뒤 추출
    # 만약 없는 경우, 빈 문자열로 대채
    if title is not None:
        row_title = title.text.replace("\n", " ")
    else:
        row_title = ""

    # 페이지 내 본문 정보에서 개행 문자를 제거한 뒤 추출
    # 만약 없는 경우, 빈 문자열로 대체
    if content_paragraphs is not None:
        for paragraphs in content_paragraphs:
            if paragraphs is not None:
                content_corpus_list.append(paragraphs.text.replace(
                    "\n", " "
                ))
            else:
                content_corpus_list.append("")
    else:
        content_corpus_list.append("")


    # 모든 정보를 하나으 데이터 프레임에 저장
    row = [row_title, "".join(content_corpus_list)]
    series = pd.Series(row, index=df.columns)
    df = df.append(series, ignore_index=True)
```
